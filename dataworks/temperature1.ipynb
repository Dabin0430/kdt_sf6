```python
# 파일입출력
import csv
import matplotlib.pyplot as plt

with open("datas/temp_seoul.csv", "r", encoding='utf-8') as f:
    data = csv.reader(f)   # 데이터 읽어온 자료 저장
    next(data) # 다음행으로 이동

    temp_high = []
    for row in data:
        if row[-1] != '':
            if int(row[0].split('-')[0]) >= 1980: #연도를 정수화 비교
            # row[-1] = float(row[-1])
                temp_high.append(float(row[-1]))

# print(len(temp_high))
# print(temp_high)
# print(max(temp_high))
plt.rc('font', family="Malgun Gothic")
plt.plot(temp_high, color='r')
plt.title("1980년 이후 서울의 최고기온")
plt.show()
```

```python
# 최고 기온(최대값) 구하기
import csv

with open("datas/temp_seoul.csv", "r", encoding='utf-8') as f:
    data = csv.reader(f)   # 데이터 읽어온 자료 저장
    next(data) # 다음행으로 이동

    max_temp = -99 #최대값 설정
    max_date = "" #최고기온 날짜
    for row in data:
        if row[-1] != '':
            row[-1] = float(row[-1])
            row[0] = row[0].replace('\t', '') #'\t'를 제거
            if row[-1] > max_temp:
                max_temp = row[-1]
                max_date = row[0]

print(f'최고기온: {max_temp}, 날짜:{max_date.strip()}')
```
